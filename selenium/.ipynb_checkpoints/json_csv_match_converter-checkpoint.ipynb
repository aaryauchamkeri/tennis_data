```python
import pandas as pd
import json
```

```python
data = []
with open('./matches.json', 'r') as file:
    data = json.load(file)
```

```python
data
```

```python
def transform(obj):
    newobj = {}
    for key in obj.keys():
        if key != 'winner' and key != 'runnerUp':
            newobj[key] = obj[key]
    
    for key in obj['winner'].keys():
        if key == 'country':
            newobj['winner_' + key] = obj['winner'][key]['name']
        else:
            newobj['winner_' + key] = obj['winner'][key]

    for key in obj['runnerUp'].keys():
        if key == 'country':
            newobj['runnerUp_' + key] = obj['runnerUp'][key]['name']
        else:
            newobj['runnerUp_' + key] = obj['runnerUp'][key]
    return newobj
```

```python
print(transform(data[0]))
```

```python
newData = [transform(obj) for obj in data]
```

```python
newData = []

for obj in data:
    try:
        newData.append(transform(obj))
    except:
        pass
```
